```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Model': [
        'Stacking Technique',
        'Bagging Technique',
        'Gradient Boosting Technique'
    ],
    'Accuracy': [0.9923, 0.9942, 0.9932],
    'Precision': [0.9915, 0.9927, 0.9925],
    'Recall': [0.9930, 0.9957, 0.9937],
    'F1 Score': [0.9923, 0.9942, 0.9931],

}

metrics_df = pd.DataFrame(data)

# Melt the DataFrame for easier plotting
metrics_df_melted = metrics_df.melt(id_vars='Model', var_name='Metric', value_name='Score')

# Set the order of models based on accuracy
order = metrics_df.sort_values('Accuracy', ascending=False)['Model'].tolist()

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Model', y='Score', hue='Metric', data=metrics_df_melted, order=order, palette="Pastel1", width = 0.5)
plt.title('Model Performance Metrics')
plt.xlabel('Model')
plt.ylabel('Score')
plt.xticks(rotation=45)
plt.ylim(0.95, 1.0)  # Set y-axis limits for better visualization
plt.legend(title='Metric', loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Data for models and time taken
data = {
    'Model': ['Stacking', 'Bagging', 'Boosting'],
    'Time Taken (seconds)': [1096.25, 391.54, 1.6653]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the plot style
sns.set(style="whitegrid")

# Create the bar plot with purple bars
plt.figure(figsize=(8, 5))
sns.barplot(x='Model', y='Time Taken (seconds)', data=df, color='#38618C', width = 0.25)

# Add labels and title
plt.title('Time Taken to Run Different Models', fontsize=14)
plt.xlabel('Comparision of Model', fontsize=12)
plt.ylabel('Time Taken (seconds)', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
